```python
exoplanet_dataset= exoplanet_dataset.drop(['P_DETECTION_MASS', 'P_GEO_ALBEDO',
'S_MAGNETIC_FIELD', 'S_DISC', 'P_ATMOSPHERE', 'P_ALT_NAMES',
'P_DETECTION_RADIUS', 'P_GEO_ALBEDO_ERROR_MIN', 'P_TEMP_MEASURED',
'P_GEO_ALBEDO_ERROR_MAX', 'P_TPERI_ERROR_MAX', 'P_TPERI_ERROR_MIN',
'P_TPERI', 'P_OMEGA_ERROR_MIN', 'P_OMEGA_ERROR_MAX', 'P_DENSITY',
'P_ESCAPE', 'P_POTENTIAL', 'P_GRAVITY', 'P_OMEGA',
'P_INCLINATION_ERROR_MAX', 'P_INCLINATION_ERROR_MIN', 'P_INCLINATION',
'P_ECCENTRICITY_ERROR_MAX', 'P_ECCENTRICITY_ERROR_MIN', 'S_TYPE',
'P_ECCENTRICITY','P_IMPACT_PARAMETER_ERROR_MIN',
'P_IMPACT_PARAMETER_ERROR_MAX', 'P_IMPACT_PARAMETER', 'P_MASS_ERROR_MAX',
'P_MASS_ERROR_MIN', 'P_HILL_SPHERE', 'P_SEMI_MAJOR_AXIS_ERROR_MIN',
'P_SEMI_MAJOR_AXIS_ERROR_MAX', 'P_MASS', 'S_AGE_ERROR_MAX',
'S_AGE_ERROR_MIN'],

axis = 1)
```

```python
exoplanet_dataset.info()
```

```python
exoplanet_dataset.isnull().sum()
```

```python
exoplanet_dataset.fillna(exoplanet_dataset.median(numeric_only = True ))
```

```python
exoplanet_dataset.info()
```

```python
object_cols = exoplanet_dataset.select_dtypes(include='object').columns

for col in object_cols:
    mode = exoplanet_dataset[col].mode
    exoplanet_dataset[col] = exoplanet_dataset[col].fillna(mode)
```

```python
float_cols = exoplanet_dataset.select_dtypes(include='float64').columns

for col in float_cols:
    mean = exoplanet_dataset[col].mean()
    exoplanet_dataset[col] = exoplanet_dataset[col].fillna(mean)
```

```python
exoplanet_dataset.info()
```

```python
object_cols = exoplanet_dataset.select_dtypes(include='object').columns

for col in object_cols:
    print(f"{col} types: {exoplanet_dataset[col].apply(type).unique()}")
```

```python
problem_cols = ['P_TYPE', 'S_TYPE_TEMP', 'P_TYPE_TEMP']

for col in problem_cols:
    exoplanet_dataset[col] = exoplanet_dataset[col].astype(str)
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoders = {}
object_cols = exoplanet_dataset.select_dtypes(include='object').columns

for col in object_cols:
    le = LabelEncoder()
    exoplanet_dataset[col] = le.fit_transform(exoplanet_dataset[col])
    label_encoders[col] = le
```

```python
exoplanet_dataset.info()
```

```python
exoplanet_dataset['P_HABITABLE'].value_counts()
```

```python
from sklearn.ensemble import RandomForestClassifier
rfs = RandomForestClassifier(class_weight='balanced')
```

```python
x = exoplanet_dataset.drop('P_HABITABLE',axis =1)
y = exoplanet_dataset['P_HABITABLE']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,stratify = y,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(
    n_estimators=100 ,
    max_depth=None ,
    random_state=42 ,
    class_weight='balanced'
)
```

```python
rfc.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
rfc_train_prediction = rfc.predict(x_train)
rfc_train_accuracy = accuracy_score(rfc_train_prediction,y_train)
rfc_train_accuracy
```

```python
rfc_test_prediction = rfc.predict(x_test)
rfc_test_accuracy = accuracy_score(rfc_test_prediction, y_test)
rfc_test_accuracy
```

```python
from sklearn.metrics import classification_report

rfc_report = classification_report(y_test, rfc_test_prediction)
rfc_report
```

```python
from sklearn.metrics import confusion_matrix ,ConfusionMatrixDisplay
cm = confusion_matrix(y_test,rfc_test_prediction)
matrix = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=rfc.classes_)
matrix.plot(cmap='Blues',values_format='d')
plt.title('Random Forest Confusion Matrix')
plt.show()
```

```python
report = classification_report(y_test,rfc_test_prediction , output_dict=True)
report_df = pd.DataFrame(report).transpose()

report_df.iloc[:-3]["f1-score"].plot(kind='bar', color='pink')
plt.title("F1-Score")
plt.ylabel("F1 Score")
plt.ylim(0, 1.1)
plt.grid(True)
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 5)
knn.fit(x_test,y_test)
```

```python
knn_train_prediction = knn.predict(x_train)
knn_train_accuracy = accuracy_score(knn_train_prediction , y_train)
knn_train_accuracy
```

```python
knn_test_prediction = knn.predict(x_test)
knn_test_accuracy = accuracy_score(knn_test_prediction, y_test)
knn_test_accuracy
```

```python
knn_report = classification_report(y_test,knn_test_prediction,zero_division=0)
knn_report
```

```python
cm = confusion_matrix(y_test,knn_test_prediction)
matrix = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=rfc.classes_)
matrix.plot(cmap='Blues',values_format='d')
plt.title(' KNN Confusion Matrix')
plt.show()
```

```python
report = classification_report(y_test,knn_test_prediction , output_dict=True, zero_division = 0)
report_df = pd.DataFrame(report).transpose()

report_df.iloc[:-3]["f1-score"].plot(kind='bar', color='pink')
plt.title("F1-Score")
plt.ylabel("F1 Score")
plt.ylim(0, 1.1)
plt.grid(True,alpha=0.5)
plt.show()
```

```python
from sklearn.ensemble import AdaBoostClassifier
ad = AdaBoostClassifier(random_state=42,algorithm='SAMME')
ad.fit(x_train,y_train)
```

```python
ad_train_prediction = ad.predict(x_train)
ad_train_accuracy = accuracy_score(ad_train_prediction , y_train)
ad_train_accuracy
```

```python
ad_test_prediction = ad.predict(x_test)
ad_test_accuracy = accuracy_score(ad_test_prediction, y_test)
ad_test_accuracy
```

```python
ad_report = classification_report(y_test,ad_test_prediction)
ad_report
```

```python
cm = confusion_matrix(y_test,ad_test_prediction)
matrix = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=rfc.classes_)
matrix.plot(cmap='Blues',values_format='d')
plt.title('Ada Boost Confusion Matrix')
plt.show()
```

```python
report = classification_report(y_test,ad_test_prediction , output_dict=True)
report_df = pd.DataFrame(report).transpose()

report_df.iloc[:-3]["f1-score"].plot(kind='bar', color='pink')
plt.title("F1-Score")
plt.ylabel("F1 Score")
plt.ylim(0, 1.1)
plt.grid(True,alpha=0.5)
plt.show()
```
